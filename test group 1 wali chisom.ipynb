```python
df = pd.read_csv(StringIO(data))
plt.figure(figsize=(16, 12))
```

```python
# 1. Age distribution
plt.subplot(3, 3, 1)
df['Age'].hist(bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
```

The church has a fairly balanced gender mix, but females slightly outnumber males.

This might reflect higher female engagement in church activities, a common trend in many congregations.

```python
# 2. Gender distribution
plt.subplot(3, 3, 2)
df['Gender'].value_counts().plot(kind='bar', color=['lightcoral', 'lightblue'])
plt.title('Gender Distribution')
plt.ylabel('Count')
```

Most members fall between 20â€“40 years, indicating a youthful congregation.

Very few members are under 20 or over 60, suggesting that targeted outreach could help attract children/teens and seniors.

```python
# 3. Ministry representation
plt.subplot(3, 3, 3)
df['Ministry'].value_counts().plot(kind='bar', color='orange')
plt.title('Ministry Representation')
plt.ylabel('Count')
```

The Choir has the largest number of members, followed by Ushering and Youth ministries.

Ministries like Media and Prayer have fewer members, which could mean a need for recruitment drives in those areas.

```python
# 4. Attendance vs Donations
plt.subplot(3, 3, 4)
plt.scatter(df['AttendanceLastMonth'], df['TotalDonations'], color='green')
plt.title('Attendance vs Donations')
plt.xlabel('Attendance (Last Month)')
plt.ylabel('Total Donations')
```

A large portion of members attend church 3â€“4 times a month, showing consistent engagement.

Thereâ€™s a small group that attends only once, possibly indicating irregular members or new joiners.

```python
# 5. Donations by ministry
plt.subplot(3, 3, 5)
df.groupby('Ministry')['TotalDonations'].sum().plot(kind='bar', color='purple')
plt.title('Total Donations by Ministry')
plt.ylabel('Total Donations ($)')
```

Donations are varied, with a small group giving significantly more than others.

Most donations are in the â‚¦50â€“â‚¦200 range, suggesting steady but modest giving habits.



```python
# 6. Membership trend
plt.subplot(3, 3, 6)
df['MembershipYear'] = pd.to_datetime(df['MembershipDate']).dt.year
df['MembershipYear'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title('New Members per Year')
plt.xlabel('Year')
plt.ylabel('Number of Members')
```

Membership increased steadily from 2010 to 2018, with a peak in 2015.

The past two years saw a slight dip, possibly due to external factors like COVID-19.

```python
# 7. Average donation by gender
plt.subplot(3, 3, 7)
df.groupby('Gender')['TotalDonations'].mean().plot(kind='bar', color='teal')
plt.title('Average Donation by Gender')
plt.ylabel('Average Donation ($)')
```

Members in the Prayer Ministry and Media Ministry contribute slightly higher amounts on average.

Youth Ministry members have lower average donations, possibly due to younger age and lower income.



```python
# 8. Age vs Donations
plt.subplot(3, 3, 8)
plt.scatter(df['Age'], df['TotalDonations'], color='red')
plt.title('Age vs Donations')
plt.xlabel('Age')
plt.ylabel('Total Donations')
```

Thereâ€™s a visible trend where more frequent attenders tend to donate more.

However, some high givers attend less often, suggesting some members support financially even without weekly presence.

```python
# 9. Attendance frequency
plt.subplot(3, 3, 9)
df['AttendanceLastMonth'].value_counts().sort_index().plot(kind='bar', color='brown')
plt.title('Attendance Frequency')
plt.xlabel('Days Attended')
plt.ylabel('Number of Members')

plt.tight_layout()
plt.show()
```

Members between 35â€“50 years tend to give the highest amounts.

Younger members (20â€“25) generally donate less, again likely due to income differences.
